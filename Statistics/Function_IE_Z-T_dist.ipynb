# Function to calculate Interval Estimate for a given set of mean,std. dev, type and sample size and to find the type of distribution

```python
import pandas as pd
from math import sqrt
```

```python
# For T distribution calculations, values of degree of freedom is also needed. hence we need T-distribution table too
```

```python
t=pd.read_excel('Student-t-distribution-table.xlsx','Sheet1')
# At 5% confidence value, α/2 = 0.025 and zα/2 = 1.96 and value of T is taken from T-table
```

```python
t
```

```python
def interval_estimate(flag,mean,sigma,sample_size):
    distribution={}
    if sample_size==0:
        print('Sample size cannot be zero, please enter correct sample value')
    else:
        if sample_size>30:
            distribution['z']=1.96
            print('Considering Z distribution for sample size greater than 30')
            if flag=='point':
                point_estimate=mean
                print("point estimate is the mean of the given sample \nand it's value is",point_estimate)
            elif flag=='interval':
                s=distribution['z']*sigma/sqrt(sample_size)
                IEU=mean+s
                IED=mean-s
                print('The interval Estimate is' , IED,'to',IEU)   
            else:
                print('Please add valid point that you want to find')
    
        elif sample_size<30:
            try:
                distribution['t']=t[0.025].loc[sample_size-1]
            except Exception as e:
                print('Sample size cannot be negative or zero')
            else:
                print('Considering T distribution for sample size less than 30')
                if flag=='point':
                    point_estimate=mean
                    print("point estimate is the mean of the given sample \nand it's value is",point_estimate)
                elif flag=='interval':
                    try:
                        s=distribution['t']*sigma/sqrt(sample_size)
                    except Exception as e:
                        print('Sample size cannot be zero or negative')
                    else:
                        IEU=mean+s
                        IED=mean-s
                        print('The Interval Estimate is' , IED,'to',IEU)   
                else:
                    print('Please add valid point that you want to find')
            
        else:
            print('At sample size of 30, both Z and T distributions show common characteristics')
            if flag=='point':
                point_estimate=mean
                print("point estimate is the mean of the given sample \nand it's value is",point_estimate)
            elif flag=='interval':
                s=distribution['z']*sigma/sqrt(sample_size)
                IEU=mean+s
                IED=mean-s
                print('The interval Estimate is' , IED,'to',IEU)   
            else:
                print('Please add valid point that you want to find')
    
        

```

```python
interval_estimate('interval',5,1,110)
```
